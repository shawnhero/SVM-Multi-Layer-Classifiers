#Predict Attributes By Landmarks

##Read the file

```python
import h5py
import numpy as np
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import average_precision_score
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from scipy import misc
%matplotlib inline
```

```python
hfile =  h5py.File('../LFW_meta.hdf5','r')
landmarks = np.transpose(hfile['landmark'])
is_train = np.array(hfile['is_train'])
attributes = np.array(hfile['attribute_annotation'])
attr_names = np.array(hfile['attribute_fields'])
filenames = np.array(hfile['img_filenames'])
rootpath = "./../LFW_image/"
```

##Select From the 73 Attributes

Analyse the prevalence of the 73 attributes, abandon highly imbalanced attributes and those making no sense (like baby and child).

```python
plt.figure(figsize=(15,45))
rows = 15
cols = 5
for i in range(73):
    plt.subplot(rows, cols, i+1)
    plt.hist(attributes[i], alpha=0.75, facecolor='g')
    plt.title(str(i)+attr_names[i])
    #non_zero_attr = attributes[i, attributes[i]!=0]
plt.savefig('attributes_analysis.png', dpi=150, bbox_inches='tight')
```

```python
attr_mask = np.ones(73, dtype=bool)
#baby
attr_mask[4] = False
#child
attr_mask[5] = False
#sunglasses
attr_mask[15] = False
#color photo
attr_mask[53] = False
#indians
attr_mask[57] = False
attributes = attributes[attr_mask,:]
attr_names = attr_names[attr_mask]
```

```python
train_lmks = landmarks[is_train==1]
test_lmks = landmarks[is_train==0]
```

```python
# for all the attributes, choose the best C value
precisions = np.zeros(attributes.shape[0])
best_c = np.zeros(attributes.shape[0])
avg_precisions = np.zeros(attributes.shape[0])
#plt.figure(figsize=(15,45))
for i in range(attributes.shape[0]):
    #print attr_names[i]
    train_attr = attributes[i,is_train==1]
    test_attr = attributes[i,is_train==0]
    
    use_train_attr = train_attr[train_attr!=0]
    use_train_lmks = train_lmks[train_attr!=0]
    
    use_test_attr = test_attr[test_attr!=0]
    use_test_lmks = test_lmks[test_attr!=0]
    
    ## build the model
    c_list = [0.1, 0.2,0.5,1.0,2.0,5.0,10.0,50.0,100.0,500.0,1000]
    precision = np.empty(len(c_list))
    for c in range(len(c_list)):
        clf = svm.SVC(C=c_list[c], class_weight ='auto')
        clf.fit(use_train_lmks, use_train_attr)
        # confidence output
        predicted_score = clf.decision_function(use_test_lmks)
        # predicted labels
        predicted_result = clf.predict(use_test_lmks)
        ## calculate the average precision
        #precision[c] = average_precision_score(use_test_attr, predicted_score)
        cmat = confusion_matrix(use_test_attr,predicted_result)
        if cmat[0,0]<cmat[1,1]:
            precision[c] = 1.0*cmat[1,1]/(cmat[0,1]+cmat[1,1])
        else:
            precision[c] = 1.0*cmat[0,0]/(cmat[0,0]+cmat[1,0])
    
    # get the best c
    avg_precisions[i] = max(precision)
    best_c[i] = c_list[np.argmax(precision)]
    plt.plot(range(len(c_list)), precision)
plt.ylabel('precision')
plt.savefig('c_value_normal_precision.png', dpi=150, bbox_inches='tight')
```

```python
# save c results
f = h5py.File('c_values.hdf5','w-')
f.create_dataset("c", data=best_c)
f.close()
```

```python
print np.mean(avg_precisions)
sort_index = np.argsort(avg_precisions)
y_pos = np.arange(len(attr_names))
plt.figure(figsize=(8, 25), dpi=200)
plt.barh(y_pos, avg_precisions[sort_index], align='center', alpha=0.4)
plt.yticks(y_pos, attr_names[sort_index])
plt.xlabel('Average Precision')
plt.title('Average Precision of the Predicted Attributes')
plt.show()
```
