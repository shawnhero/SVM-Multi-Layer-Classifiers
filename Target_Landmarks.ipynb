```python
import h5py
import numpy as np
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import average_precision_score
import matplotlib.pyplot as plt
from scipy import misc
import colorsys
%matplotlib inline
```

```python
hfile =  h5py.File('../LFW_meta.hdf5','r')
landmarks = np.transpose(hfile['landmark'])
landmarks = landmarks.reshape(landmarks.shape[0], 49,2)
filenames = np.array(hfile['img_filenames'])
rootpath = "./../LFW_image/"
```

```python
i = 2
face = misc.imread(rootpath+filenames[i])
plt.figure(figsize=(10, 10), dpi=200)
plt.imshow(face)
plt.scatter(landmarks[i][:, 0], landmarks[i][:, 1], c='r', marker='+', s=100)
for j in range(49):
    plt.annotate(j, (landmarks[i][j, 0], landmarks[i][j, 1]))
# draw the face color area
upleft = np.array([landmarks[i,1,0], landmarks[i,11,1]], dtype=int)
downright = np.array([landmarks[i,4,0], landmarks[i,16,1]], dtype=int)
for x in range(upleft[0], downright[0]+1):
    for y in range(upleft[1], downright[1]+1):
        if (x==upleft[0] or y==upleft[1] or x==downright[0] or y==downright[1]):
            plt.scatter(x, y, c='b', marker='_')


# draw the hair color area
delta_y = landmarks[i,41,1] - landmarks[i,0,1]
upleft = np.array([landmarks[i,1,0], landmarks[i,1,1]-delta_y], dtype=int)
downright = np.array([landmarks[i,9,0], landmarks[i,1,1]-delta_y+3], dtype=int)
for x in range(upleft[0], downright[0]+1):
    for y in range(upleft[1], downright[1]+1):
        if (x==upleft[0] or y==upleft[1] or x==downright[0] or y==downright[1]):
            plt.scatter(x, y, c='w', marker='_')

plt.savefig('landmark_position_label.png', dpi=150, bbox_inches='tight')
```

#Add Hair Color and Skin Color Attributes

```python
#up_left(x_1, y_11), down_right(x_4, y_16)
def mean_rgb(upleft, downright, image):
    rgbs = []
    for i in range(upleft[0], downright[0]+1):
        for j in range(upleft[1], downright[1]+1):
            rgbs.append(image[i,j])
    rgbs = np.array(rgbs)
    return np.mean(rgbs, axis=0)

def mean_rgb_face(landmarks, image):
    face_upleft = np.array([landmarks[1,0], landmarks[11,1]], dtype=int)
    face_downright = np.array([landmarks[4,0], landmarks[16,1]], dtype=int)
    return mean_rgb(face_upleft, face_downright, image)

def mean_rgb_hair(landmarks, image):
    delta_y = landmarks[41,1] - landmarks[0,1]
    hair_upleft = np.array([landmarks[1,0], landmarks[1,1]-delta_y], dtype=int)
    hair_downright = np.array([landmarks[9,0], landmarks[1,1]-delta_y+3], dtype=int)
    return mean_rgb(hair_upleft, hair_downright, image)
```
